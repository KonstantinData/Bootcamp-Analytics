# Group by and Aggregation

```python
import pandas as pd

flav = pd.read_csv(r"D:\Repositories\Git_GitHub\PandasAlexTheAnalyst\Flavors.csv")
flav
```

```python
# Gruppieren nach 'Base Flavor' und den Mittelwert berechnen
grouped = flav.groupby('Base Flavor').mean(numeric_only=True)

print(grouped)
```

```python
flav.groupby("Base Flavor").count()
```

```python
flav.groupby("Base Flavor").min()
```

```python
flav.groupby("Base Flavor").max()
```

```python
flav.groupby("Base Flavor").sum(numeric_only=True)
```

```python
flav.groupby("Base Flavor").agg({"Flavor Rating":["mean","max","count","sum"]})
```

```python
flav.groupby("Base Flavor").agg({"Flavor Rating":["mean","max","count","sum"],"Texture Rating":["mean","max","count","sum"]})
```

```python
flav.groupby(["Base Flavor","Liked"]).mean(numeric_only=True)
```

```python
flav.groupby(["Base Flavor","Liked"]).agg({"Flavor Rating":["mean","max","count","sum"]})
```

```python
flav.groupby(["Base Flavor"]).describe()
```

# Merge, Join, and Concatenate

```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"D:\Repositories\Git_GitHub\PandasAlexTheAnalyst\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"D:\Repositories\Git_GitHub\PandasAlexTheAnalyst\LOTR 2.csv")
df2
```

## Merge

```python
df1.merge(df2, how= "inner")
```

```python
df1.merge(df2, how= "outer")
```

```python
df1.merge(df2, how= "left")
```

```python
df1.merge(df2, how= "right")
```

```python
df1.merge(df2, how= "cross")
```

## Join

```python
df1.join(df2, on="FellowshipID", how="outer", lsuffix="_Left", rsuffix="_Right")
```

```python
df4 = df1.set_index("FellowshipID").join(df2.set_index("FellowshipID"), lsuffix="_Left", rsuffix="_Right", how="outer")
df4
```

## Concatenate

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], join="inner")
```

```python
pd.concat([df1, df2], join="outer")
```

```python
pd.concat([df1, df2], join="outer", axis=1)
```
